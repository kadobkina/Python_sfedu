```python
import numpy as np
import pandas as pd
```

**Создание объекта Series**
(одномерная последовательность с индексом)

```python
# 4. Создайте объект Series из сгенерированного случайного массива numpy.
si=pd.Series(np.random.randint(1, 11, 5))
si
```

```python
# 5. Выдать на экран представление данных массива (values) и его индекса (index).
# Обратиться по индексу к элементу Series и поменять его значение. 
# Присвойте имя объекту Series (свойство name), и его индексу (index.name). 
# Выведите его тип значений (dtype).
print(f'Данные - {si.values}, индексы - {si.index}')
print(f'Старое значение 1-го элемента массива(si[0]) - {si[0]}')
si[0] = si[0]*-1
print(f'Новое значение 1-го элемента массива(si[0]) - {si[0]}')
si.name = "Integers"
print(f'Имя объекта - {si.name}')
si.index.name = "Indexes"
print(f'Имя индекса - {si.index.name}')
print(f'Тип значений - {si.dtype}')
```

**Создание объекта DataFrame**
(таблица с заголовками столбцов и с индексом по строкам)

```python
# 9. Создайте объект DataFrame из словаря объектов Series. 
# Используйте данные в разных объектах Series разных типов. Например, в первом – целые числа, во втором – строки, и т.д.
ints_dict = {'0': 5,
             '1': 6,
             '2': 7,
             '3': 8,
             '4': 9}
ints = pd.Series(ints_dict)

strings_dict = {'0': 'one',
             '1': 'two',
             '2': 'three',
             '3': 'four',
             '4': 'five'}
strings = pd.Series(strings_dict)

floats_dict = {'0': 0.,
             '1': 0.1,
             '2': 0.2,
             '3': 0.3,
             '4': 0.4}
floats = pd.Series(floats_dict)

df = pd.DataFrame({'ints': ints,
                   'strings': strings,
                   'floats': floats})
df
```

```python
# 10. Выдать на экран представление данных в виде матрицы numpy (values), его индекса (index), столбцов (columns) и типы столбцов (dtypes).
print(f'Данные в виде матрицы - \n{df.values}')
print(f'Индексы - {df.index}')
print(f'Столбцы - {df.columns}')
print(f'Типы столбцов - \n{df.dtypes}')
```

```python
# 11. Выдайте на экран информацию о созданном выше объект DataFrame: размерность (ndim, shape), количество элементов (size).
print(f'Размерность - {df.ndim} {df.shape}') # ndim - Return an int representing the number of axes / array dimensions. Return 1 if Series. Otherwise return 2 if DataFrame
print(f'Количество элементов - {df.size}')
```

```python
# 12. Обратиться к одному столбцу объект DataFrame: имя[имя столбца]. 
# Выдать на экран представление данных этого столбца (values) и его индекса (index). 
# Выведите на экран его имя (name) и его тип значений (dtype).
print(f'Столбец ints - \n{df["ints"]}')
print(f'Данные столбца ints - {df["ints"].values}')
print(f'Данные индекса столбца ints - {df["ints"].index}')
print(f'Имя столбца ints - {df["ints"].name}')
print(f'Тип значений столбца ints - {df["ints"].dtype}')
```
